```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Blood Type']=le.fit_transform(df['Blood Type'])
df['Medication']=le.fit_transform(df['Medication'])
df['Medical Condition']=le.fit_transform(df['Medical Condition'])
df['Test Results']=le.fit_transform(df['Test Results'])
df['Doctor']=le.fit_transform(df['Doctor'])
df['Hospital']=le.fit_transform(df['Hospital'])
df
```

```python
a = df.drop(['Name', 'Date of Admission', 'Doctor', 'Hospital', 'Insurance Provider', 'Admission Type', 'Discharge Date','Billing Amount','Room Number', 'Test Results'], axis=1)
a
```

```python
x=df.drop('Test Results',axis=1)
y=df['Test Results']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
from sklearn.svm import SVC

training_features = ['Age', 'Gender', 'Blood Type', 'Medical Condition', 'Medication']

# Ensure x_train and x_test contain only the desired training features
x_train_filtered = x_train[training_features]
x_test_filtered = x_test[training_features]

# Re-initialize and re-train the model with the correct features
model = SVC()
model.fit(x_train_filtered, y_train)

# Now predict with the consistently filtered x_test
y_pred = model.predict(x_test_filtered)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
