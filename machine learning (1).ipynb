```python
#Linear Regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
data = {
    'Experience': [1, 2, 3, 4],
    'Salary': [3, 4, 5, 6]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Features and target
X = df[['Experience']]
y = df['Salary']
```

```python
# Model creation
model = LinearRegression()
model.fit(X, y)
```

```python
new_data = pd.DataFrame({'Experience': [5]})
prediction = model.predict(new_data)
print("Predicted Salary:", prediction)
```

```python
plt.figure()
plt.scatter(df['Experience'], df['Salary'])
plt.plot(df['Experience'], model.predict(X))
plt.xlabel("Years of Experience")
plt.ylabel("Salary (LPA)")
plt.title("Linear Regression: Salary vs Experience")
plt.show()
```

```python
#Multilinear Regression
# Create dataset
data = {
    'Area': [1000, 1500, 2000, 2500],
    'Bedrooms': [2, 3, 3, 4],
    'Age': [10, 5, 2, 1],
    'Price': [50, 75, 95, 130]
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['Area', 'Bedrooms', 'Age']]
y = df['Price']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
new_house = pd.DataFrame({
    'Area': [1800],
    'Bedrooms': [3],
    'Age': [4]
})
```

```python
predicted_price = model.predict(new_house)
print("Predicted House Price:", predicted_price)
```

```python
#Area vs Price
plt.figure()
plt.scatter(df['Area'], y)
plt.plot(df['Area'], model.predict(X))
plt.xlabel("Area (sqft)")
plt.ylabel("Price (Lakhs)")
plt.title("Area vs House Price")
plt.show()
```

```python

#Bedrooms vs Price
plt.figure()
plt.scatter(df['Bedrooms'], y)
plt.plot(df['Bedrooms'], model.predict(X))
plt.xlabel("Bedrooms")
plt.ylabel("Price (Lakhs)")
plt.title("Bedrooms vs House Price")
plt.show()
```

```python
# Age vs Price
plt.figure()
plt.scatter(df['Age'], y)
plt.plot(df['Age'], model.predict(X))
plt.xlabel("Age of House (Years)")
plt.ylabel("Price (Lakhs)")
plt.title("Age vs House Price")
plt.show()
```

```python
#Logistic Regression
# Sample dataset
data = {
    'Email_Length': [100, 300, 150, 400],
    'Links': [1, 5, 0, 7],
    'Spam': [0, 1, 0, 1]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Features and target
X = df[['Email_Length', 'Links']]
y = df['Spam']
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
new_email = pd.DataFrame({
    'Email_Length': [200],
    'Links': [2]
})
```

```python
prediction = model.predict(new_email)
print("Spam or Not (1=Spam, 0=Not Spam):", prediction)
```

```python
# Use Email_Length for visualization
X_email = df[['Email_Length']]
y_spam = df['Spam']
```

```python
# Train a logistic model with one feature (for plotting)
model_1d = LogisticRegression()
model_1d.fit(X_email, y_spam)
```

```python

# Create smooth values for curve
x_range = np.linspace(
    X_email.min().values[0],
    X_email.max().values[0],
    300
)

# Convert to DataFrame with correct feature name
x_range_df = pd.DataFrame({'Email_Length': x_range})

# Predict probabilities
y_prob = model_1d.predict_proba(x_range_df)[:, 1]
```

```python
# Plot
plt.figure()
plt.scatter(df['Email_Length'], y_spam)
plt.plot(x_range, y_prob)
plt.xlabel("Email Length")
plt.ylabel("Probability of Spam")
plt.title("Logistic Regression (Sigmoid Curve)")
plt.show()
```
