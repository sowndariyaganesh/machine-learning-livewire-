```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Property_Area']=le.fit_transform(data['Property_Area'])
data
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
data['ApplicantIncome']=ss.fit_transform(data[['ApplicantIncome']])
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

x = data.drop(['LoanAmount','Loan_Amount_Term','Property_Area'], axis=1)
y = data['Property_Area']


x = x.drop('Loan_ID', axis=1)

x['Dependents'] = x['Dependents'].replace('3+', '3').astype(float)
x['Dependents'] = x['Dependents'].fillna(x['Dependents'].mode()[0])


categorical_cols = x.select_dtypes(include='object').columns
x = pd.get_dummies(x, columns=categorical_cols, drop_first=True)

numerical_cols_with_nans = x.select_dtypes(include=np.number).isnull().sum()
numerical_cols_with_nans = numerical_cols_with_nans[numerical_cols_with_nans > 0].index
for col in numerical_cols_with_nans:
    x[col] = x[col].fillna(x[col].mode()[0])


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


model = RandomForestClassifier(random_state=42)
model.fit(x_train, y_train)
```
