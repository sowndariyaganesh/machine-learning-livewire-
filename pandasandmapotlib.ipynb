```python
data['Age']=data['Age'].fillna(data['Age'].median())
data
```

```python
data['Cabin']=data['Cabin'].fillna('Unknown')
data
```

```python
data.isnull().sum()
```

```python
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])
```

```python
data = data.drop(columns=["Cabin"])
```

```python
data["FamilySize"] = data["SibSp"] + data["Parch"] + 1
data["IsAlone"]    = (data["FamilySize"] == 1).astype(int)
data["AgeGroup"]   = pd.cut(data["Age"], bins=[0,18,60,100],
                           labels=["child","adult","senior"])
print("\nNew columns added: FamilySize, IsAlone, AgeGroup")
print(data[["FamilySize","IsAlone","AgeGroup"]].head())
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Titanic EDA — Day 4 Project", fontsize=14)

# Plot 1: Survival count
data["Survived"].value_counts().plot(kind="bar", ax=axes[0,0],
    color=["#E24B4A","#1D9E75"], title="Survived (0=No, 1=Yes)")

# Plot 2: Survival rate by gender
data.groupby("Sex")["Survived"].mean().plot(kind="bar", ax=axes[0,1],
    color=["#534AB7","#1D9E75"], title="Survival rate by gender")

# Plot 3: Age distribution
data["Age"].plot(kind="hist", bins=30, ax=axes[1,0],
    color="#BA7517", title="Age distribution")
data.groupby("Pclass")["Survived"].mean().plot(kind="bar", ax=axes[1,1],
    color="#185FA5", title="Survival rate by ticket class")

plt.tight_layout()
plt.savefig("titanic_eda.png", dpi=120)
plt.show()
print("Chart saved as titanic_eda.png")
```
